Quick modeling!  First, get columns ready.  Dummify 'species' and drop it, along with 'target':

```python
#  dummify 'species' column & get rid of original

big_df = pd.concat([df.drop(['species', 'target'], axis = 1), \
            pd.get_dummies(df['species'], drop_first = True)], axis = 1)

big_df.head()
```

Notice normalize = 'True' argument will automatically normalize the features for us, and fit_intecept = 'True' will take care of the intercept ("constant") term, so no need to add a column of 1's:

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings(action="ignore", module="scipy", message="^internal gelsd")

lm = LinearRegression(normalize = True, fit_intercept = True)

X = big_df[['sepal_L', 'petal_L', 'petal_W', 'versicolor', 'virginica']]

y = big_df['sepal_W']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 121)

lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
y_pred = lm.predict(X_test)
lm_r2 = r2_score(y_test, y_pred)

lm_r2
```

```python
target_species = pd.DataFrame(list(df.ix[y_test.index]['species']))

target_species.columns = ['species']

target_colors = target_species['species'].apply(lambda x: species_dict[x])
```

```python
plt.figure(figsize=(12,8))

plt.scatter(y_test, y_pred, s = 40, alpha = 0.6, color = target_colors)
plt.title('Iris Regression: Actual vs Predicted Sepal Length', fontsize = 24)

plt.xlim([1.8, 4.5])
plt.ylim([1.8, 4.5])

plt.ylabel('Predicted', fontsize = 20)
plt.xlabel('Actual', fontsize = 20)

plt.plot([1.0, 5.0], [1.0, 5.0], '--', linewidth = 0.5, color = 'red', alpha = 1)

plt.text(2, 4, 'R^2 = %s' %lm_r2.round(2), fontsize = 20)
plt.show()
```

Let's visualize relative sizes of our regression coefficients:

```python
coefs = pd.DataFrame(lm.coef_, index = X.columns)
coefs.columns = ['coef']
coefs['coef'] = coefs['coef'].apply(lambda x: abs(x))
coefs = coefs.sort_values(by = 'coef', ascending = False)

coefs
```

Here, we've made a dataframe with feature in the index and absolute value of corresponding regression coefficient in the column and then sorted it in descending order.  

```python
plt.figure(figsize=(12,8))

N = 5
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

plt.bar(ind, abs(coefs['coef']), width)

plt.ylabel('Coefficient Magnitude', fontsize = 20)
plt.title('Sepal Width Regressors: Magnitude of Coefficient', fontsize = 24, y = 1.03)
plt.xticks(ind + 0.18, coefs.index, fontsize = 14)
plt.xlim([-0.5, 5])
plt.ylim([0, coefs['coef'][0] + 0.2])
#plt.yticks(np.arange(-1.8, 0.2, 0.2), fontsize = 14)
plt.xlabel('Feature', fontsize = 20)

plt.show()
```

```python
lm_little = LinearRegression(fit_intercept = True, normalize = True)

lm_little.fit(X_train[['virginica', 'versicolor', 'petal_W']], y_train)
```

```python
lm_little.coef_
```

```python
lm_little.intercept_
```

```python
y_little_pred = lm_little.predict(X_test[['virginica', 'versicolor', 'petal_W']])

lm_little_r2 = r2_score(y_test, y_little_pred)

lm_little_r2
```

```python
plt.figure(figsize=(12,8))

plt.scatter(y_test, y_little_pred, s = 40, alpha = 0.6, color = target_colors)
plt.title('Iris Regression: Actual vs Predicted Sepal Length', fontsize = 24)

plt.xlim([1.8, 4.5])
plt.ylim([1.8, 4.5])

plt.ylabel('Predicted', fontsize = 20)
plt.xlabel('Actual', fontsize = 20)

plt.plot([1.0, 5.0], [1.0, 5.0], '--', linewidth = 0.5, color = 'red', alpha = 1)
plt.text(2, 4, 'R^2 = %s' %lm_little_r2.round(2), fontsize = 20)

plt.show()
```

Clearly, these models are predicting too large on smaller sepals and too small on larger ones.  Next steps might be to figure out why this is happening.  Add other two features back in, investigate outliers more closely, try ridge and/or lasso, ... try anything & figure out how to justify it!
