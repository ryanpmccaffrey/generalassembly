## Changing Time Frequencies

```python
goog[['Close']]
```

## Using agg on resample

```python
goog[['Close']].resample('M').agg(['min', 'max', 'mean', len])
```

```python
goog[['Close']].resample('Q').ohlc()
```

## Adjusting the days

```python
goog = web.DataReader('GOOG', 'yahoo', start, end)
goog.head()
```

```python
# adding a day and week ago's closes
goog['Prior Day Close'] = goog['Close'].shift(1)
goog['Prior Week Close'] = goog['Close'].shift(7)
```

```python
goog.head(10)
```

```python
goog['Closed Up - 1 Day'] = np.where(goog['Close'] > goog['Close'].shift(1), 1, 0)
goog['Closed Up - 1 Week'] = np.where(goog['Close'] > goog['Close'].shift(7), 1, 0)
```

```python
# here we select only columns with the word 'Close' in them
# we would also only want to use rows without nans (but left here to illustrate - see next slide)
goog[[x for x in goog.columns if 'Close' in x and 'Adj' not in x]]
```

```python
goog[[x for x in goog.columns if 'Close' in x and 'Adj' not in x]].dropna()
```

## Conclusion:
 - Time series data is simply data that is ordered by date
 - Care must be used when testing the data - usually done in sliding windows (train first 3 months, test next 3 -> train first 3 shifted 1 day
forward day, test next 3 plus one day forward, and so on)
 - pandas has extensive built-in functionality for working with time series data: pandas_datareader, resample, asfreq, shift, etc.
 - The DateTimeIndex is the key to this time series functionality
